### How all the stemming are working

```python
ow=[] #original_words
ps=[] #PorterStemmer
ls=[] #LancasterStemmer
ss=[] #SnowballStemmer
wl=[] #WordNetLemmatizer
for i in data1['reviews']:
    for word in word_tokenize(i):
        ow.append(word)
        ps.append(PorterStemmer().stem(word))
        ls.append(LancasterStemmer().stem(word))
        ss.append(SnowballStemmer('english').stem(word))
        wl.append(WordNetLemmatizer().lemmatize(word))
```

```python
pd.DataFrame({'original_words':ow,'PorterStemmer':ps,'LancasterStemmer':ls,'SnowballStemmer':ss,'WordNetLemmatizer':wl})
```
